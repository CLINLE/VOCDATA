```python
import pandas as pd
from pathlib import Path

excel_path = Path(
    r"C:\Users\claud\iCloudDrive\Dokumente\02_CLI\Studium\ZHAW\Masterarbeit\vocdata\data\bfs_data.xlsx"
)

xls = pd.ExcelFile(excel_path)

for sh in xls.sheet_names:
    if sh.lower().endswith("_data"):
        # Header-Zeile ermitteln: erste Zeile, in der mindestens 3 Zellen nicht leer sind
        top = pd.read_excel(xls, sheet_name=sh, nrows=15, header=None)
        header_row = next(
            i for i, row in top.iterrows() if row.notna().sum() >= 3
        )

        df = pd.read_excel(xls, sheet_name=sh, header=header_row)
        print("=" * 80)
        print(f"▶︎ Sheet: {sh}")
        display(df.head(10))        # zeigt sauber formatiert im Notebook

```
