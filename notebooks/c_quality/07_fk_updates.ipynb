```python
# ------------------------------------------------------------
# Fehlende Dimensionstabellen direkt per SQLAlchemy anlegen
# ------------------------------------------------------------
from sqlalchemy import create_engine, text

engine = create_engine(
    "mysql+pymysql://root:voc_root@localhost:3306/vocdata",
    echo=False, future=True
)

ddl_script = """
CREATE TABLE IF NOT EXISTS dim_gemeindetyp (
    gemeindetyp_id   INT AUTO_INCREMENT PRIMARY KEY,
    gemeindetyp_code VARCHAR(50) UNIQUE,
    gemeindetyp_bez  VARCHAR(100)
);
CREATE TABLE IF NOT EXISTS dim_sprachregion (
    sprachregion_id   INT AUTO_INCREMENT PRIMARY KEY,
    sprachregion_code VARCHAR(50) UNIQUE,
    sprachregion_bez  VARCHAR(100)
);
CREATE TABLE IF NOT EXISTS dim_kanton (
    kanton_id   INT AUTO_INCREMENT PRIMARY KEY,
    kanton_code CHAR(2) UNIQUE,
    kanton_bez  VARCHAR(100)
);
"""

with engine.begin() as con:
    for stmt in ddl_script.strip().split(");"):
        stmt = stmt.strip()
        if stmt:
            con.execute(text(stmt + ");"))

print("✔ Drei leere Dimensionstabellen angelegt.")
```

```python
# ------------------------------------------------------------
# Kantons­codes + Langbezeichnungen einmalig in dim_kanton laden
# ------------------------------------------------------------
kanton_rows = [
    ("AG", "Aargau"),      ("AI", "Appenzell Innerrhoden"),
    ("AR", "Appenzell Ausserrhoden"),        ("BE", "Bern"),
    ("BL", "Basel-Landschaft"),              ("BS", "Basel-Stadt"),
    ("FR", "Freiburg"),    ("GE", "Genf"),
    ("GL", "Glarus"),      ("GR", "Graubünden"),
    ("JU", "Jura"),        ("LU", "Luzern"),
    ("NE", "Neuenburg"),   ("NW", "Nidwalden"),
    ("OW", "Obwalden"),    ("SG", "St. Gallen"),
    ("SH", "Schaffhausen"),("SO", "Solothurn"),
    ("SZ", "Schwyz"),      ("TG", "Thurgau"),
    ("TI", "Tessin"),      ("UR", "Uri"),
    ("VD", "Waadt"),       ("VS", "Wallis"),
    ("ZG", "Zug"),         ("ZH", "Zürich")
]

from sqlalchemy import create_engine, text
engine = create_engine(
    "mysql+pymysql://root:voc_root@localhost:3306/vocdata",
    echo=False, future=True
)

with engine.begin() as con:
    # leere Tabelle vorher bereinigen (nur einmal nötig)
    con.execute(text("TRUNCATE TABLE dim_kanton;"))
    con.execute(
        text("""
        INSERT INTO dim_kanton (kanton_code, kanton_bez)
        VALUES (:code, :bez)
        """),
        [{"code": c, "bez": b} for c, b in kanton_rows]
    )

print("✔ dim_kanton mit 26 Einträgen befüllt.")
```
