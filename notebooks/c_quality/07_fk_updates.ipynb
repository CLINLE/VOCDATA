```python
# ------------------------------------------------------------
# Fehlende Dimensionstabellen direkt per SQLAlchemy anlegen
# ------------------------------------------------------------
from sqlalchemy import create_engine, text

engine = create_engine(
    "mysql+pymysql://root:voc_root@localhost:3306/vocdata",
    echo=False, future=True
)

ddl_script = """
CREATE TABLE IF NOT EXISTS dim_gemeindetyp (
    gemeindetyp_id   INT AUTO_INCREMENT PRIMARY KEY,
    gemeindetyp_code VARCHAR(50) UNIQUE,
    gemeindetyp_bez  VARCHAR(100)
);
CREATE TABLE IF NOT EXISTS dim_sprachregion (
    sprachregion_id   INT AUTO_INCREMENT PRIMARY KEY,
    sprachregion_code VARCHAR(50) UNIQUE,
    sprachregion_bez  VARCHAR(100)
);
CREATE TABLE IF NOT EXISTS dim_kanton (
    kanton_id   INT AUTO_INCREMENT PRIMARY KEY,
    kanton_code CHAR(2) UNIQUE,
    kanton_bez  VARCHAR(100)
);
"""

with engine.begin() as con:
    for stmt in ddl_script.strip().split(");"):
        stmt = stmt.strip()
        if stmt:
            con.execute(text(stmt + ");"))

print("✔ Drei leere Dimensionstabellen angelegt.")
```

```python
# ## Data Quality Check: Anzahl Null-Fremdschlüssel pro Dimension
# %%  ─────────  LVA-Faktentabelle grob prüfen  ─────────
import pandas as pd
from sqlalchemy import create_engine, text

engine = create_engine(
    "mysql+pymysql://root:voc_root@localhost:3306/vocdata",
    echo=False, future=True
)

qry = """
SELECT
    COUNT(*)                                    AS total_rows,
    SUM(abschlussniveau_id  IS NULL)            AS null_abschlussniveau,
    SUM(lernform_id         IS NULL)            AS null_lernform,
    SUM(geschlecht_id       IS NULL)            AS null_geschlecht,
    SUM(mig_status_id       IS NULL)            AS null_mig_status,
    SUM(anschlussart_id     IS NULL)            AS null_anschlussart,
    SUM(qv_status_id        IS NULL)            AS null_qv_status,
    SUM(isced_id            IS NULL)            AS null_isced,
    SUM(beruf_id            IS NULL)            AS null_beruf
FROM fact_lva_stats;
"""

with engine.begin() as con:
    res = con.execute(text(qry)).mappings().one()

print("Null-Fremdschlüssel in fact_lva_stats:")
for k, v in res.items():
    print(f"  {k}: {v}")
```

```python
# %% ---------- Vollständigkeits-Audit fact_lva_stats ----------
import pandas as pd
from sqlalchemy import text

qry = """
SELECT
  COUNT(*)                                      AS total_rows,

  -- Kennzahlen auf NULL prüfen
  SUM(anzahl_lernende_lva            IS NULL)   AS null_lernende_lva,
  SUM(anzahl_lernende_wiedereinstieg IS NULL)   AS null_wiedereinstieg,
  SUM(anzahl_lernende               IS NULL)    AS null_lernende,
  SUM(anzahl_lehrvertraege_lva       IS NULL)   AS null_lehrvertraege_lva,
  SUM(anzahl_lehrvertraege           IS NULL)   AS null_lehrvertraege,

  -- FK-IDs auf 0 (UNKNOWN) prüfen
  SUM(abschlussniveau_id   = 0) AS fk_abschl_niv,
  SUM(lernform_id          = 0) AS fk_lernform,
  SUM(geschlecht_id        = 0) AS fk_geschlecht,
  SUM(mig_status_id        = 0) AS fk_mig_status,
  SUM(anschlussart_id      = 0) AS fk_anschlussart,
  SUM(qv_status_id         = 0) AS fk_qv_status,
  SUM(lva_zeitraum_id      = 0) AS fk_lva_zeitraum,
  SUM(wiedereinst_dauer_id = 0) AS fk_wieder_dauer,
  SUM(isced_id             = 0) AS fk_isced,
  SUM(beruf_id             = 0) AS fk_beruf
FROM fact_lva_stats;
"""

audit = pd.read_sql(text(qry), engine)
display(audit.T)   # transponiert = übersichtlicher
```
